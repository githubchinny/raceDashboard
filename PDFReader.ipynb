```python
import PyPDF2
import os
import pandas as pd

```

```python
from sys import platform
# if platform == "linux" or platform == "linux2":
    # linux    
if platform == "darwin":
    # OS X
    dir = "/Users/mark/Documents/GitHub/raceDashboard/Race Dashboard data/"
elif platform == "win32":
    # Windows...
    sanofi_share = "C:/Users/mark_/Sanofi/Sanofi x McLaren sharing - General/"
    folder = "Frankfurt sprint/SFD/Alarms_data/AL6_Etikettierer/"

pd.set_option('display.max_columns', None)
pd.set_option('max_colwidth', None)
```

```python
# Get all the PDF filenames.
pdfFiles = []
pages = pd.DataFrame()

for filename in os.listdir( sanofi_share + folder ):
    if filename.endswith('.pdf'):
        pdfFiles.append(filename)

pdfFiles.sort(key=str.lower)

# debugging - use just one file
# pdfFiles = ['2021-08-12 103619 ReportReview.pdf']

# Loop through all the PDF files.
for filename in pdfFiles:
    pdfFileObj = open(sanofi_share + folder + filename, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

    # Loop through all the pages (except the first) and add them.
    for pageNum in range(0, pdfReader.numPages):
        print (filename, pageNum)
        pageObj = pdfReader.getPage(pageNum)
        page = pageObj.extractText()
        # split each row on '0000' which is the row number
        page_df = pd.DataFrame(page.split('0000'))
        # ignore the first row after the splitting as it's blank
        page_df = page_df[1:]
        # split each row on ';' and label the columns
        df2 = pd.DataFrame(page_df[0].str.split(';').tolist(),
            columns = ['Nummer','Projekt','Status','Prio','Quali','Datum','Zeit','UTC Versatz','Meldetext','ID','BMK','Parameterwert','Einheit','Alt/Neu-Wert','Benutzer','Auftrag','Charge'])
        pages = pages.append(df2)


```

```python
pages['Datetime'] = pd.to_datetime(pages.Datum + ' ' + pages.Zeit)
pages.head()
```
