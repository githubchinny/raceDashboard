```python
import pandas as pd
```

```python
dir = "C:/Users/mark_/McLaren Technology Group/McLaren Accelerator - Sanofi - Sanofi/Data Analysis/"
# output_dir = "C:/Users/mark_/Documents/McLaren2021/Sanofi/"
output_dir = "C:/Users/mark_/McLaren Technology Group/McLaren Accelerator - Sanofi - Sanofi/Data Analysis/"

# Extract these files from QlikSense for AL6, Apr + May + Jun, from the ROOT loss analysis page (using the scatter chart)
file = (dir + 'Buffer Times.xlsx')
df_min = pd.read_excel(file, sheet_name='M&E Duration')
df_OEE = pd.read_excel(file, sheet_name='M&E OEE')
```

```python
df = df_min.merge(df_OEE[['Reason Machine', 'Point of OEE (%)']], on='Reason Machine')
```

```python
# round to 2 decimal places 
df[['Duration (min)','Point of OEE (%)']] = round(df[['Duration (min)','Point of OEE (%)']],2)
```

```python
df[['Reason', 'Machine']] = df_min['Reason Machine'].str.split(' - ', 1, expand=True)
```

```python
df.head(50)
```

```python
groupby_machine = df.groupby(['Machine'])[['Point of OEE (%)','Duration (min)']].sum().sort_values(by='Point of OEE (%)', ascending=False)
groupby_machine
```

```python
groupby_reason = df.groupby(['Machine','Reason'])[['Point of OEE (%)','Duration (min)']].sum().sort_values(by='Machine', ascending=False)
groupby_reason.head(15)
```

```python
path = r"C:/Users/mark_/McLaren Technology Group/McLaren Accelerator - Sanofi - Sanofi/Data Analysis/bufferCalcs.xlsx"
writer = pd.ExcelWriter(path, engine='openpyxl')
groupby_machine.to_excel(writer, 'Machine')
groupby_reason.to_excel(writer, 'Reasons')
writer.save()
```
